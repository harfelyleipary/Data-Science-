```python
import tkinter as tk
from tkinter import ttk

class FTI_UKSW_App(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("FTI UKSW Application")
        self.geometry("700x500")

        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        self.frames = {}
        for F in (HomePage, AboutPage):
            frame = F(container, self)
            self.frames[F] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame(HomePage)

        nav_frame = tk.Frame(self, bg="lightgray")
        nav_frame.pack(side="top", fill="x", padx=10, pady=5)
        ttk.Button(nav_frame, text="Home FTI UKSW",
                   command=lambda: self.show_frame(HomePage)).pack(side="left", padx=5)
        ttk.Button(nav_frame, text="About FTI UKSW",
                   command=lambda: self.show_frame(AboutPage)).pack(side="left", padx=5)

    def show_frame(self, cont):
        self.frames[cont].tkraise()

class HomePage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        ttk.Label(self, text="Welcome to Home FTI UKSW!", font=('Arial', 18, 'bold')).pack(pady=20)
        ttk.Label(self, text="This is the home page for FTI UKSW. \n\nHere you can find general information, news, and updates.",
                  wraplength=600, justify="center", font=('Arial', 12)).pack(pady=10)

class AboutPage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        ttk.Label(self, text="About FTI UKSW", font=('Arial', 18, 'bold')).pack(pady=20)
        ttk.Label(self, text="The Faculty of Information Technology (FTI) at UKSW provides quality education in IT. \n\nOur vision is to be a leading faculty in information technology.",
                  wraplength=600, justify="left", font=('Arial', 12)).pack(pady=10)

if __name__ == "__main__":
    app = FTI_UKSW_App()
    app.mainloop()
```

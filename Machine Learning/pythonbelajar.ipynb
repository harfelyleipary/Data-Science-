<h4>Class&Instance Variables</h4>

```python
class Herro:
    jumlah = 0 #Static
    def __init__(self, 
                 inputName,
                 inputPower,
                 inputArmor):
        #Instance Variable
        self.name = inputName
        self.power = inputPower
        self.armor = inputArmor
        Herro.jumlah += -2
        print("Membuat Hero dengan nama " + inputName)


hero1 = Herro("Jak", 10,4);
print(Herro.jumlah)
hero2 = Herro("Nonde", 20, 5);
print(Herro.jumlah)
hero3 = Herro("Petric", 89, 10);
print(Herro.jumlah)

#print(Herro.__dict__) #Nanti ngak ada. ada variable statict. variabel yang ada dalam
#variabel itu sendiri
```

<h3>Methods</h3>

```python
#yang kita lakukan terhadap object. atau si object bisa berinteraksi dengan object yang lain.
#Object ada 2 yaitu : 
#1.Method interative :dengan client(pribadi)
#2.Interaksi antar object

class Hero:
    #class variabel
    jumlah_hero = 0;

    def __init__(self, 
                 inputName, 
                 inputHealth, 
                 inputPower, inputArmor):
        self.name = inputName
        self.health = inputHealth
        self.power = inputPower
        self.armor = inputArmor
        Hero.jumlah_hero += 1

    #Void function, method tanpa return 
    def siapa(self):
        print("NamaKu adalah "+self.name)

    #Method dengan argumen
    def healtUp(self,up):
        self.health +=up

hero1 = Hero('sniper', 100,10,5)
hero2 = Hero('Jaka',100,5,10) 

hero2.siapa()
hero2.healtUp(100)

print(hero2.health)

print(hero1.__dict__)
print(hero2.__dict__)
```
